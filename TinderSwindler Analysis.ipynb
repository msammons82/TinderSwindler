```python
import numpy as np
import pandas as pd

```

```python
df = pd.read_csv('TinderSwindlerVersion2.csv')
df.head(20)
```

```python
df.source.value_counts()
```

```python
df.isnull().sum()
```

```python
df1 = df.dropna(subset=['user_location'])
```

```python
df1.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,10))
sns.set(style='darkgrid')
ax = sns.countplot(y="user_location", data=df1, order=df1['user_location'].value_counts().index[0:20])
```

```python
plt.figure(figsize=(15,10))
sns.set(style="darkgrid")
ax = sns.countplot(x="source", data=df1, order=df1['source'].value_counts().index[0:5])
```

# Exploratory Data Analysis

```python
df = pd.DataFrame(df1[['user_name', 'text','user_followers']])
```

```python
plt.figure(figsize=(15,10))
sns.set(style="darkgrid")
ax = sns.countplot(x=df, data=df, order=df[df].value_counts().index[1])
```
